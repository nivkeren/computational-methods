1. Import the `pyplot` and `numpy` libraries.

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [15, 6.67]
```

2. write a function that receives a polynomial object (numpy poly1d) and returns 2 lists one of the real and and of the complex roots of the polynomial (the list might be empty.

```python
def real_comp_roots(poly: np.poly1d) -> (list, list):
    # YOUR CODE GOES HERE
    real_roots = [root for root in poly.roots if not np.iscomplex(root)]
    complex_roots = [root for root in poly.roots if np.iscomplex(root)]
    return real_roots, complex_roots        
```

----
From now on, we are now going to work on the following polyinomial function.

```python
poly = np.poly1d([1, -4, 0, 0, 1])
```

3. plot the polynomial function at the range of x values that will include all of its real roots.

```python
real, comp = real_comp_roots(poly)
x = np.linspace(real[0].real - 6, real[-1].real + 4, 1000)
plt.plot(x, poly(x))
plt.grid()
plt.show()
```

4. Find the extremum point points of the function

```python
der1 = poly.deriv()
der2 = der1.deriv()
extremums = [ex for ex in der1.roots if ex not in der2.roots] 
extremums
```

5. Calculate the area under the function, between the extremum to the largest root

```python
integral = poly.integ()
integral(real[0].real) - integral(extremums[0])
```

6. Plot the area.

```python
real, comp = real_comp_roots(poly)
x = np.linspace(real[0].real - 6, real[-1].real + 4, 1000)
x1 = np.linspace(real[0].real, extremums[0], 1000)
plt.plot(x, poly(x))
plt.grid()

plt.fill_between(x1, 0, poly(x1), color="C1")
plt.show()
```
