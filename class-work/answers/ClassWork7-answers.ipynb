Plot the next 4 graphs on a 2 by 2 subplots.  
**Dont forget to add axis labels and titles.**
1. plot the black contours of the function *z = 10 - 2x - y* for *-3 <= x, y <= 3* at `axes[0, 0]` add labels.
2. plot a 3d scatter plot of the function *z = 1 / (1 + x^2 + y^2) for *-2 <= x, y <= 2* at `axes[0, 1]` add a vivid colormap.
3. plot the function *z = cos(x^2 + y)* for *-2 <= x, y <= 2* at `axes[1, 0]`.
4. plot the function *z = cos(x^2 + y^2)* for 0 <= r <= 6 , -0.8 * pi <= theta <= 0.7 * pi at `axes[1, 1]` with summer colormap.

```python
ax1 = plt.subplot(221)
ax2 = plt.subplot(222, projection="3d")
ax3 = plt.subplot(223, projection="3d")
ax4 = plt.subplot(224, projection="3d")

# 1
x = np.linspace(-3, 3)
y = np.linspace(-3, 3)
xx, yy = np.meshgrid(x, y)
zz = 10 - 2 * xx - yy
cs = ax1.contour(xx, yy, zz, colors='k')
ax1.clabel(cs, fontsize=9, inline=1)

#2

x = np.linspace(-2, 2)
y = np.linspace(-2, 2)
xx, yy = np.meshgrid(x, y)
zz = 1 / (1 + xx**2 + yy**2)
cs = ax2.scatter(xx, yy, zz)

#3
zz = np.sin(xx**2 + yy)
cs = ax3.plot_surface(xx, yy, zz, shade=True)

#4
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.8 * np.pi, 0.7 * np.pi, 40)
rr, theta2 = np.meshgrid(r, theta)

xx = rr * np.sin(theta2)
yy = rr * np.cos(theta2)
zz = np.sin(np.sqrt(xx ** 2 + yy ** 2))

cs = ax4.plot_surface(xx, yy, zz, rstride=1, cstride=1,
                      cmap='summer', edgecolor='none')
```
